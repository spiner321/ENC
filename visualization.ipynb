# # Load the ENC File
---

```python
from osgeo import ogr

# Initialize the S-57 driver
driver = ogr.GetDriverByName('S57')

# Path to the ENC file
enc_file_path = '/home/kimgh/workspace/ENC/data/ENC_ROOT/EA200004.000'

# Open the data source
data_source = driver.Open(enc_file_path, 0)
```

# # Inspect Layers and Attributes
---

```python
layer_count = data_source.GetLayerCount()
for i in range(layer_count):
    layer = data_source.GetLayerByIndex(i)
    print(f'Layer {i+1}: {layer.GetName()}')
```

# # Display the Chart
---

```python
import matplotlib.pyplot as plt

# Create a new plot
plt.figure(figsize=(10, 8))

# Loop through all features in the layer
layer_count = data_source.GetLayerCount()
for i in range(layer_count):
    # Choose a layer to plot
    layer = data_source.GetLayerByIndex(i)
    for feature in layer:
        geometry = feature.GetGeometryRef()
        if geometry is not None:
            x = [geometry.GetX(i) for i in range(geometry.GetPointCount())]
            y = [geometry.GetY(i) for i in range(geometry.GetPointCount())]
            plt.plot(x, y)

# Show the plot
plt.show()
```
